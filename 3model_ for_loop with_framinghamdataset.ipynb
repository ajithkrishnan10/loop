```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/datasets/framingham.csv')
df
```

```python
df.isnull().sum()
```

```python
df['education'].fillna(df['education'].mode()[0],inplace=True)
df['cigsPerDay'].fillna(df['cigsPerDay'].mode()[0],inplace=True)
df['BPMeds'].fillna(df['BPMeds'].mode()[0],inplace=True)
df['BPMeds'].fillna(df['BPMeds'].mode()[0],inplace=True)
df['heartRate'].fillna(df['heartRate'].mode()[0],inplace=True)
df['glucose'].fillna(df['glucose'].mode()[0],inplace=True)
df['totChol'].fillna(df['totChol'].mode()[0],inplace=True)
df['BMI']=df['BMI'].fillna(df['BMI'].mean())
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
```

```python
knn=KNeighborsClassifier()
sv=SVC()
nb=GaussianNB()

lst=[sv,knn,nb]
for i in lst:
  print('**',i,'**')
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

**over** **sampiling**

```python
from imblearn.over_sampling import SMOTE
oversample=SMOTE(random_state=1)
X_os,y_os=oversample.fit_resample(X,y)
```

```python
y_os.value_counts()
```

```python
X_train_os,X_test_os,y_train_os,y_test_os=train_test_split(X_os,y_os,test_size=0.3,random_state=1)
```

```python
scaler=MinMaxScaler()
X_train_os=scaler.fit_transform(X_train_os)
X_test_os=scaler.transform(X_test_os)
```

```python
knn_os=KNeighborsClassifier()
sv_os=SVC()
nb_os=GaussianNB()

lst=[sv_os,knn_os,nb_os]
for i in lst:
  print('**',i,'**')
  i.fit(X_train_os,y_train_os)
  y_pred_os=i.predict(X_test_os)
  print(classification_report(y_test_os,y_pred_os))
```

**under sampiling**

```python
from imblearn.under_sampling import RandomUnderSampler
undersampler=RandomUnderSampler(random_state=1)
X_us,y_us=undersampler.fit_resample(X,y)
```

```python
y_us.value_counts()
```

```python
X_train_us,X_test_us,y_train_us,y_test_us=train_test_split(X_us,y_us,test_size=0.3,random_state=1)
```

```python
scaler_us=MinMaxScaler()
X_train_us=scaler_us.fit_transform(X_train_us)
X_test_us=scaler_us.transform(X_test_us)
```

```python
knn_us=KNeighborsClassifier()
sv_us=SVC()
nb_us=GaussianNB()

lst=[sv_us,knn_us,nb_us]
for i in lst:
  print('**',i,'**')
  i.fit(X_train_us,y_train_us)
  y_pred_us=i.predict(X_test_us)
  print(classification_report(y_test_us,y_pred_us))
```
